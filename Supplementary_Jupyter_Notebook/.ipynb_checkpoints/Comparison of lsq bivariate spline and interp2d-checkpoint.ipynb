```python
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt
```

```python
def fun(x, y):
    rep = 30
    return (np.sin(rep*x) + np.sin(rep*y)/5)
```

```python
numK = 10
xKnot = np.linspace(0, 1, numK)
yKnot = np.linspace(0, 1, numK)

numP=numK
x = np.linspace(0, 1, numP)
y = np.linspace(0, 1, numP)
x, y = np.meshgrid(x, y)
x = x.ravel()
y = y.ravel()
z = fun(x, y)

lsq = interpolate.LSQBivariateSpline(x, y, z, xKnot, yKnot)


plt.scatter(x, y, c = z)
plt.colorbar()
plt.show()

interp = 80
xInterp = np.linspace(0, 1, interp)
yInterp = np.linspace(0, 1, interp)
xInterp, yInterp = np.meshgrid(xInterp, yInterp)
zInterp = lsq.ev(xInterp, yInterp)
```

```python
# plt.scatter(xInterp, yInterp, c = zInterp)
plt.imshow(zInterp, origin = 'lower', extent=(0, 1, 0, 1))
plt.colorbar()
plt.show()
```

```python
numP = 10
x = np.linspace(0, 1, numP)
y = x.copy()

xx, yy = np.meshgrid(x, y)

z = fun( xx, yy )

ob = interpolate.RectBivariateSpline(x, y, z)

interp = 70
xInterp = np.linspace(0, 1, interp)
yInterp = np.linspace(0, 1, interp)
xxInterp, yyInterp = np.meshgrid(xInterp, yInterp)

zzInterp = ob(xInterp, yInterp)
zzInterp.size
```

```python
plt.scatter(xx, yy, c=z)
plt.colorbar()



plt.show()

plt.imshow(zzInterp, origin='lower')
plt.colorbar()
plt.show()

```
