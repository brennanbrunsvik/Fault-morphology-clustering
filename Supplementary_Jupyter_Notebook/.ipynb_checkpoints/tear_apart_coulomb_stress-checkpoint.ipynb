```python
#stars show locations of simulated rupture
#Figure 10 modified
#dCS vs location
def dCSVsLocation_oneplane(dCSIM, high, low, latIM, lonIM, strainers):
    matplotlib.rcParams.update({'font.size': 14})
    plt.figure(figsize=(10,10))
    
    color = []
    for i in dCS:
        if i > high:
            color.append('red')
        elif i < low:
            color.append('blue')
        else:
            color.append('gray')
    
    size = 5
    plt.scatter(lonFault, latFault, c = color, s = size)
    plt.scatter(lonFault[strainers], latFault[strainers])#, c = color, s = 500, marker = '*')

    plt.ylabel('Latitude')
    plt.xlabel('Longitude')
    plt.title('Rupture locations and positive, negative, or low magnitude $\Delta$CFF')
    #plt.savefig('Rupture locations and positive negative or low magnitude dcff', dpi = 400)
    plt.show()

high = .005e6
low = -.005e6

dCSVsLocation_oneplane(dCSMHom, high, low, latFault, lonFault, strainers)
```
