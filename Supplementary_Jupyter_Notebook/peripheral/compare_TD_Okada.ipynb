```python
lambdaLame = 26000000000.0
shearMod = 26000000000.0
poisson = lambdaLame / (2 * (lambdaLame + shearMod) )
alpha = (lambdaLame + shearMod) / (lambdaLame + 2 * shearMod)
```

```python
xcenter = 0.
ycenter = 0.
zcenter = -1000.

cellwidth   = 100.
cellheight = 100.

xleft     = xcenter - .5 * cellwidth
xright    = xcenter + .5 * cellwidth
yforward  = 0.
yback     = 0.
ztop      = zcenter + .5 * cellheight
zbottom   = zcenter - .5 * cellheight
```

```python
strikeDeg = 90.
dipDeg = 45.
rakeDeg = 180
strike = strikeDeg * np.pi / 180
dip = dipDeg * np.pi / 180
rake = rakeDeg * np.pi / 180



slip = 1.
ss = np.cos(rake) * slip
ds = np.sin(rake) * slip
ts = 0.

normalUnit, strikeUnit = faultUnitVectors(strike, dip, 0)
alongUnit = np.cross(normalUnit, strikeUnit)
```

```python
%matplotlib inline
pcenter = np.array([xcenter, ycenter, zcenter])

trilins = 2

interval = np.linspace(-.5, .5, trilins)

p = np.zeros( (trilins, trilins, 3) )
p[:, :] = pcenter.copy()

for i in range(trilins): #strike
    for j in range(trilins): #vertical
        p[i, j] += (interval[i] * strikeUnit * cellwidth
                + interval[j] * alongUnit * cellheight)

xt = p[:, :, 0].ravel()
yt = p[:, :, 1].ravel()
zt = p[:, :, 2].ravel()

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_aspect('equal')
ax.scatter(xt, yt, zt)
plt.show()
```

```python
xpoints = 56
ypoints = 56
gridwidth = 10000
xp = np.linspace(-gridwidth, gridwidth, xpoints)
yp = np.linspace(-gridwidth, gridwidth, ypoints)
xp, yp = np.meshgrid(xp, yp)
xp = xp.ravel()
yp = yp.ravel()
zp = zcenter * np.ones(xpoints * ypoints)
```

```python
def sortCounterClockwise(x, y):
    xmean = np.average(x)
    ymean = np.average(y)
    xprime = x - xmean
    yprime = y - ymean

    xhat = np.array([1, 0, 0])
    
    theta = np.zeros(x.size)

    for i in range(x.size):
        theta[i] = np.arctan2(yprime[i], xprime[i])

    argsort = np.argsort(theta)
        
    return argsort
```

```python
tri = Delaunay(np.array([xt, zt]).T)
simplices = tri.simplices
triangles = simplices.shape[0]
DisplTD = np.zeros((xpoints * ypoints, 3), dtype = np.float64)
DisplTemp = np.zeros(3, dtype = np.float64)
StressTD = np.zeros((xpoints * ypoints, 6), dtype = np.float64)
StressTemp = np.zeros(6, dtype = np.float64)
StrainTD = np.zeros((xpoints * ypoints, 6), dtype = np.float64)
StrainTemp = np.zeros(6, dtype = np.float64)
for triangle in range(triangles):
    triInds = simplices[triangle]
    simpx = xt    [triInds]
    simpy = yt    [triInds]
    simpz = zt    [triInds]
    p123 = np.array([simpx, simpy, simpz])
    P1 = p123[:, 0].copy()
    P2 = p123[:, 1].copy()
    P3 = p123[:, 2].copy()
    for point in range(xpoints * ypoints):
        P1 = p123[:, 0].copy()
        P2 = p123[:, 1].copy()
        P3 = p123[:, 2].copy()
        TDDispl.py_DispInHalfSpace(DisplTemp,
                xp.copy()[point], yp.copy()[point], zp.copy()[point],
                P1.copy(), P2.copy(), P3.copy(),
                copy.deepcopy(ss), copy.deepcopy(ds), copy.deepcopy(ts), 
                copy.deepcopy(poisson) )
        DisplTD[point] += DisplTemp
        TDStrain.py_StrainInHalfSpace(StressTemp, StrainTemp,
                xp.copy()[point], yp.copy()[point], zp.copy()[point],
                P1.copy(), P2.copy(), P3.copy(),
                copy.deepcopy(ss), copy.deepcopy(ds), copy.deepcopy(ts), 
                shearMod, lambdaLame )
        StressTD[point] += StressTemp
        StrainTD[point] += StrainTemp
```
